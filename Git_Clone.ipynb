#git クローン
IntelliJ IDEAを使用すると、既存のリポジトリをチェックアウト(Git用語ではクローン)して、ダウンロードしたデータに基づいて新しいプロジェクトを作成できる。  
1. Gitリポジトリのクローン作成の開始  
 - バージョン管理統合が有効になっている場合  
 →Git | クローン に進む
 - バージョン管理統合が有効になっていない場合  
 →VCS | バージョン管理から取得 または ファイル | 新規 | バージョン管理からプロジェクトに進む  
 - 現在開いているプロジェクトがない場合  
 →ウェルカム画面でVCSから取得をクリック
2. バージョン管理から取得ダイアログで、クローンを作成するリモートリポジトリの URL を指定するか、VCS ホスティングサービスの1つを選択する。  
3. クローンをクリックする。複製したソースに基づいてプロジェクトを作成する場合は、「はい」をクリックする。Git ルートマッピングは、プロジェクトのルートディレクトリに自動的に設定される。
4. プロジェクト'<project_name>'を信頼して開きますか？ と聞かれるので、プロジェクトセキュリティダイアログで、プロジェクトを開く方法を選択する。
5. プロジェクトを初めてインポートまたは複製すると、IntelliJ IDEA がプロジェクトを分析する。IDE が複数の構成を検出すると、使用する構成を選択するように求められる。
