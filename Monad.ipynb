###「モナドは単なる自己関手の圏におけるモノイド対象だよ。何か問題でも？」
####自己関手の圏…
関手（functor）＝合成しても圏の構造を保ち続ける写像のようなもの  
関手の合成をすることで関手圏D^Cができあがる  
関手圏には、恒等関手が存在し、結合律を満たす  
Cという関手同士を合成することで、自己関手の圏ができあがる  
自己関手の圏には、恒等関手（単位元）が存在し、関手の合成（二項演算）が成り立つ  
<br>

####モノイド対象…
モノイド  
-二項演算  
-単位元の存在  
-結合律の成立  
（モノイドの例）　（Int, +, 0）  
　　　　　　　　　（Int, *, 1）   
　　　　　　　　　（Array, +, []）  
<br>
圏論におけるモノイド  
デカルト積でペアを作って圏論におけるモノイドを構成する  

モノイド対象  
二項演算と単位射が定義でき、その射の組のことを言う  
M×M　→M←　１  
　　　↑　↑  
二項演算　単位射  
<br>
モノイダル圏  
デカルト積をテンソル積に一般化し、一元集合も一般化することでモノイダル圏が構成できる  
モノイダル圏では以下が成り立つ  
-テンソル積　C×C→C  
-単位対象　　I∊０ｂ(Ｃ)  
-結合子　　　(A⊗B)⊗C≅A⊗(B⊗C)  
-左単位子　　I⊗A≅A  
-右単位子　　A⊗I≅A  
結合子、左単位子、右単位子で等号が成立するとき、特別にストリクトモノイダル圏という  
<br>
自己関手の圏では、関手の合成（二項演算＝テンソル積）と恒等関手（単位元＝単位対象）が存在する  
二項演算では、結合律（結合子）、単位律（左単位子、右単位子）が成り立つ  
自己関手の圏＝ストリクトモノイダル圏
<br>
モナド（T,μ,η）  
-自己関手　T ：C→C  
-自然変換　μ：T⊗T→T  
-自然変換　η：I→T  
モナド則  
-結合律　　μ◦Tμ＝μ◦μT  
-単位律　　μ◦Tη＝μ◦ηT＝１T  
モナド＝自己関手の圏（ストリクトモノイダル圏）におけるモノイド対象
<br>

クライスリトリプル  
モナドの自然変換μの代わりに拡張演算=<<を用いる  
-自己関手　T ：C→C  
-自然変換　η：I→T  
-拡張演算　=<<：(A → T(B)) → T(A) → T(B)  

クライスリ合成  
μC◦T(g)◦f ＝ ｇ<=<f
