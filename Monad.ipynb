##「モナドは単なる自己関手の圏におけるモノイド対象だよ。何か問題でも？」
###自己関手の圏…
関手（functor）＝合成しても圏の構造を保ち続ける写像のようなもの  
関手の合成をすることで関手圏D^Cができあがる  
関手圏には、恒等関手が存在し、結合律を満たす  
Cという関手同士を合成することで、自己関手の圏ができあがる  
自己関手の圏には、恒等関手（単位元）が存在し、関手の合成（二項演算）が成り立つ  
<br>

####モノイダル圏  
デカルト積をテンソル積に一般化し、一元集合も一般化することでモノイダル圏が構成できる  
モノイダル圏では以下が成り立つ  
- テンソル積　C×C→C  
- 単位対象　　I∊０ｂ(Ｃ)  
- 結合子　　　(A⊗B)⊗C≅A⊗(B⊗C)  
- 左単位子　　I⊗A≅A  
- 右単位子　　A⊗I≅A  

結合子、左単位子、右単位子で等号が成立するとき、特別にストリクトモノイダル圏という  
<br>
自己関手の圏では、関手の合成（二項演算＝テンソル積）と恒等関手（単位元＝単位対象）が存在する  
二項演算では、結合律（結合子）、単位律（左単位子、右単位子）が成り立つ  
自己関手の圏＝ストリクトモノイダル圏
<br>

#圏
圏 C は以下のものからなる：  
- 対象の類 　ob(C)  
- 対象の間の射の類　 hom(C)  

各射 f ∈ hom(C) には始域と呼ばれる対象 a ∈ ob(C) および終域と呼ばれる対象 b ∈ ob(C) が付随して、「f は a から b への射である」 と言い、＜f: a → b＞ と書き表す。  
a から b への射の類 (hom-class; ホム類) hom(a, b) は a から b への射全体の成す類を言う  

このとき、任意の三対象 a, b, c ∈ ob(C) に対し、射の合成と呼ばれる二項演算
＜hom(a, b) × hom(b, c) → hom(a, c)；(f, g) ↦ g ∘ f＞ が存在して以下の公理を満たす

結合律：  
＜f: a → b＞, ＜g: b → c＞, ＜h: c → d＞ ならば h ∘ (g ∘ f) = (h ∘ g) ∘ f が成り立つ  
単位律：  
各対象 x ∈ ob(C) に対して x の恒等射と呼ばれる自己射 ＜idx = 1x: x → x＞ が存在して、任意の射 ＜f: a → x＞,＜g: x → b＞ に対して ＜1x ∘ f = f＞、＜g ∘ 1x = g＞ を満たす  

#関手
2 つの圏 C, D があったとき、

C の対象 X に対し D の対象 F(X) を与える  
射 f : X → Y に対し、射 F(f) : F(X) → F(Y) を与える  
このとき以下の性質を満たす  
- 各対象 X ∈ C に対して F(idX) = idF(X)   
- 任意の射 f: X → Y および g: Y → Z に対して F(g ∘ f) = F(g) ∘ F(f)  

このFを関手という  
※関手に対して恒等射および射の合成を保存する

#自然変換
自然変換 (natural transformation) は 2 つの関手間の関係である。
C,Dを圏とする   
関手 ＜F:C→D＞ から 関手 ＜G:C→D＞ への変換 ＜μ:F⇒G＞ を考える  
C の対象でパラメータ付けられた D の射の族 {μX: F(X) → G(X)}X ∈ C であって、任意の C の射 ＜f : X → Y＞ に対して、　μY◦F(f)=G(f)◦μX　を満たすとき、＜μ:F⇒G＞を自然変換という

#クライスリ・トリプル
圏 C 上のクライスリ・トリプル（Kleisli triple）とは、自己関手 ＜T : C → C＞、自然変換 μ（μX : X → TX , X ∈ Obj(C)）、射 f : X → TY に対して射 f* : TX → TY を与える拡張演算子（extension operator） _* からなる三つ組 （T, μ, _*）で、以下を満たすものを言う  
- ηX* = 1TX
- f* ◦ μX = f 　ただし、f : X → TY
- g* ◦ f* = (g* ◦ f)* 　ただし、f : X → TY かつ g : Y → TC

#モナド
Cを圏とする  
C上のモナドは、自己関手 ＜T ：C→C＞と２つの自然変換 ＜μ：T⊗T→T＞、＜η：I→T＞からなり、＜T, μ, η＞と表す  
モナドはモナド則という次の条件を満たす  
- μ◦Tμ＝μ◦μT  
- μ◦Tη＝μ◦ηT＝I

#モナドとクライスリ・トリプルの対応づけ
クライスリトリプルの拡張演算子 _* はモナドを用いて定義できる  
- モナド→クライスリトリプル  

任意の射 f:X→TY に対して、拡張演算子 _* を f* = μY◦Tf とすればよい  
- クライスリトリプル→モナド  

拡張演算子 _* から自然変換 μ:T⊗T→T をつくる  
射の属 μ＝{idTA*:TTA→TA | A∊Obj(c)} が自然変換 μ:T⊗T→T となる
