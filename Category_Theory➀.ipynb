#Programming
###一番最低水準のアセンブリ言語のプログラミング
コンピュータが正確に何をするかを書き、メモリからもってきて、それをレジスタに入れて、アドレスとして使い、それでジャンプとかそんなことをする。
これはすごく命令的なプログラミングである。
そして、このプログラミングはチューリング機械と関係している。
チューリング機械というのは、とても原始的な機械で、紙のテープに、スタンプ的なものを押すだけのもので、ただのアセンブリ言語みたいに、この数字を読んで、それをテープに戻して、テープから何か消してという作業をするだけのもの。
これがプログラミングへのひとつのアプローチであるのだが、これらは全部、コンピュータができる前に発明されている。
###プログラミングへのアプローチ＜ラムダ計算＞
それで、数学から出てきた別のアプローチが、アロンゾ・チャーチが創案したラムダ計算である。
それは、「計算可能性とはなにか？」みたいなものである。
数学について考えるのだが、使う言葉は、「実行」と「変換」である。
だが、これらのアプローチはあまり実用的でないため、より高水準の抽象化をもたらす言語が出てきた。
###プログラミングへのアプローチ＜手続き型プログラミング＞
それが、「手続き型プログラミング」である。
この「手続き型プログラミング」の特徴というのは、大きな問題を「手続き」に分ける。
それぞれの「手続き」には「名前」があり、決まった数の「引数」があって、ときどき「値」を返す。
やることを細かくぶった切るから、より大きい問題を扱うことができる。
###プログラミングへのアプローチ＜オブジェクト指向プログラミング＞
次に人々が思いついたのは、「オブジェクト指向プログラミング」である。
これは、さらに抽象的なものである。
ブツはオブジェクトの中に隠してしまって、オブジェクトを合成できる。
ただオブジェクトの表面だけ、つまり、インターフェースだけ見て、中身を見ずにオブジェクトを結合することができる。
オブジェクト同士を結合することで、より大きな問題を扱える。
より大きな問題を扱いたければ、大きい問題を小さい問題にぶった切ればよい。
それらを、別々に解いて、それらの解を結合するのだ。
この動作には、コンポーザビリティ（合成可能性）という名前がついている。
###プログラミングへのアプローチ＜抽象化＞
他にも、プログラミングに役立つものがある。
それは、抽象化である。
抽象化（abstraction）はギリシャ語から来ていて、意味は演算（subtraction）と似ている。
詳細は気にせず、同じものであるように扱う。
オブジェクト指向プログラミングでは、「オブジェクト」がこの詳細を隠すなにかだ。
それでさらに、抽象データ型なんてものまであって、こいつらはインターフェースだけ露出してて、どうやって実装されてるかは知らなくていいことになってる。
この事物を抽象化することと、組み合わせることを一緒にすると、リユーザビリティ（再利用可能性）というアイディアになる。

#ObjectOriented Programming
ここで、「オブジェクト指向のアプローチ」の何かがおかしいことに気づく。
それは、並行性は「オブジェクト指向プログラミング」とうまく混ざらないということだ。
なぜなら、オブジェクトは実装を隠す、まさしくダメなものを隠すからだ。
このせいで、合成可能にならない。
オブジェクトはとても重要なものを2つ隠す。
それは、変化（mutation）と共有（sharing）だ。
変化と共有を混ぜることを、データ競合という。
つまり、「オブジェクト指向プログラミング」のオブジェクトが覆い隠して抽象化しているのは、データ競合だ。
だが、データ競合は覆い隠して抽象化していいものではない。
ただ言いたいのは、この種の抽象化の水準を上げるときには、なにを覆い隠して抽象化しているのか気をつけなければならないということだ。
実はその前からあったのだが、次の抽象化は「関数型プログラミング」である。
関数型言語には変化がない。
だから、データ競合が隠されるなんて問題はない。
それに、データ構造をより大きなデータ構造に組み合わせる方法もある。
これもすべて、Immutable（変化不能）なおかげだ。
だからものを安全に組み合わせたり、分解したりできる。

#Template Metaprogramming
C++で使える、最も高水準の抽象化はテンプレートメタプログラミングだ。
これらはとても単純なアイデアだが、この言語がこいつらを表現するのにややこし過ぎるだけなんだ。
だが、このめちゃくちゃ複雑で巨大なテンプレート は、ハスケルで一行じゃないかと気づく。
抽象化の水準を飛び越えるような、より高水準の抽象度でのプログラミングをだい ぶ簡単にする、そんな言語が存在する。
テンプレートメタプログラミングを使えば、 保持する型が違うだけの、 大量のデータ構造を作るというアイデアが表現できる。
例えば、 integerのvectorと、booleanのvectorみたいな。
共有できるコードが大量に存在する。

#Abstraction in Haskell
ハスケルの中でも抽象化の限界があることに気づく。
残念ながら、 ハスケルで表現するのがちょっぴりややこしい、 とても抽象的なことって存在する。
そこで、カテゴリ論だ。
だが、カテゴリ論からハスケルに物事を翻訳する時は、たくさんの抽象度を失う。
その物事をより具体的にするということだ。
カテゴリ論はハスケルよりさらに高い水準の言語だ。
これは実用的にコードがかける言語ではない。
ハスケルから学んだから、ハスケルでの構成はどうなってるかとか、 どうやってことを成し遂げるか変化不能のデー 夕構造でアルゴリズムをどうやって実装するかとかを知ってる。
カテゴリ論からこれらのめちゃくちゃ抽象的なアイデアを取ってきて、ハスケルっていうややこしい言語に翻訳すればいい。
これらのアイデアが、カテゴリ論か らハスケルへと浸透していくと、そいつらはさらに、C++やアセンブリへ、PHPへ、JavaScriptへでさえも浸透していく。
これらはとても普遍的なアイディアだからだ。

#Theoretical Motivation
最も高い水準の抽象度に到達したい。
後でプログラムに翻訳できるアイディアを表現するためにだ。
もし、十分抽象化すれば、エベレストの頂上から見下ろしているように、全てのものが同じように見え始める。
この高い高い水準で は突然、他のものまで同じに見える。
数学者の人らはこれを発見したんだ。数学では、個別に領域を発展させてきた。
最初に発展したのは地測学（幾何学）、代数、 数論、位相論、そして集合論だ。
これらは全部、全く違う別の理論で、何も似てない。
カテゴリ論はこれら全部の類似性を発見したんだ。
つまり、ある特定の水準の抽象度では、これら全部の理論の構造は同じなのだ。
つまり統合方法があるのだ。
さらにカテゴリ論には、プログラミングもある。メモリーとか、プロセッサとか、レジスタとかがついたコンピュータを扱うプログラミングだ。
プログラミング言語もあるってこと だ。
こいつは数学的に記述できる。
そう、ラムダ計算だ。
基本的にこれらの言語は全部、 ラムダ計算にルーツがあるんだ。
それから、カテゴリ論にはデータ構造もある。
これは、配管工がパイプを弄ってるみたいに、 低水準だ。
そしたら、 コンピュータ・サイエンスの人らは、「いや、データ構造は型を形成する。つまり、型理論というのがある。」というのだ。
これら全てのデータ構造のカテゴリを記述できる型理論があるんだ。

#Type Theory
しかし、数学にも型理論が存在してて、数学で人々が型理論を発明したのは、コンピュータ・サイエンスの人らが解こうとし てる問題を解くためじゃなくて、パラドックスとかの問題を解こうとするためなのだ。
「ラッセルのパラドックス(全ての集合の集合は作れない)」や、「散髪屋のパラドックス(散髪屋は自分のヒゲを剃ることができるのか?)」みたいなものを。
実は、この散髪屋のパラドックスはラッセルのパラドックスに翻訳できる。
それでラッセルはこの型理論を思いついたんだ。
「集合の集まりは、階層上の階層を形成する。全ての集合を直接一つの集合に入れることはできない。」
それで、 型理論はここから発展して、とても抽象的なマーティン=レーフ型理論ってのがあって、こいつはとても形式的だ。
つまり、型理論はパラドックスを扱うための数学の一系統に過ぎないってことだ。

#Abstract Thinking
人類は進化を10億年やり続けてるんだ。
だが一方で、ここ1万年くらいかそこらで進化したやつもあるんだ。
進化からしたら最後の1ミリ秒だ。俺らは突然この脳を持った。
この脳は抽象的に考えられるんだ。数を数えることができるし、コミュニケーションできるし、何かを整理したりもできる。
さらには、数学もできるし、科学もできる。これはめちゃくちゃ新しい能力だ。
人類はここ数千年、科学してるが、 こんなの進化のスケールからしたら無に等しい。
一体なんでこの脳が出てきたのか?
マンモスを狩ったり、サーベルタイガーから逃げたりするためだ。
プログラミングするために進化した んじゃない。
自分らを群れとして編制するため、コミュニケーションのため、ようは社会活動のためだ。
だから俺らは言語能力を持ってるんだ。
抽象的に考えたり言語を使う能力 は、新しい進化だ。
今はこのツールを使って数学するんだ。
複雑な状況をより単純なものに分解して、そいつらを解くんだ。
これはどこにでも現れる。
これはあまりにも全てに浸透しすぎていて、 気づきすらしないんだ。
しかし、これが俺達の仕組みだ。これが俺達の仕組みなのだから、 科学の仕組みもこれと同じだ。
つまり、全ての科学の系統と、 全ての数学の系統は、ぶった切って、また一緒にできるものだけなのだ。
だから、こいつらが同じに見えるのは不思議じゃない。
構造を持った問題だけしか人類には見えないからだ。

#Renormalization
そして、構造をぶった切っていけば、点にたどり着くはずだと考えるのが普通だろう。
しかし、点だと仮定しても上手くいかず、ひもだとしてもダメで、量子に至っては、合成ができない。
そこで思い至るのだ。
ひょっとしてひょっとすると、一番底 では、多分底なんて無いんだけど、 とても底の方のものは分けられないのかもしれない。多分、ものは合成可能じゃないんだ。この合成可能性は、多分、自然の属性では無いのだ、と。
多分これは単に、脳の属性なんだ。
おそらく人類の脳がこんなだから、至るところに構造を見つけるのだ。
そしてもし構造を見つけられないと、ただ諦めるんだ。
こう考えると、カテゴリ論てのは実は、 数学とか物理学の話じゃないんだ。
カテゴリ論とは脳はどういう仕組みかっていう思考の話なんだ。
どっちかというと、実在論っていうより認識論なんだ。
認識論ってのは、 「どうやって俺らが物事を推論するか」とか、 「どうやって物事を学べるか」とかだ。
実在論ってのは、「ものとは何か」だ。多分俺らは、「ものとは何か」は学べないんだ。
だが俺らは、「どうやって学べるか」を学ぶことができる。
これが、カテゴリ論の教えだ。

#Overview
ここまでをまとめよう。
ここまで話してきたのは、カテゴリ論とハスケルそしてプログラミングについてだ。
###カテゴリ論について
カテゴリ論というのは、ほぼ最も抽象的な数学の分野だ。
そして、思考の話でもあり、認識論なのだ。
カテゴリ論は実は、プログラミングに役立つ。プログラミングへのアプローチは抽象度が高いほうが良いのだが、カテゴリ論はハスケルよりも高い水準の言語だからだ。
###ハスケル
ハスケルというのは、純粋関数型プログラミング言語のことである。
これは非常に扱うことが難しい言語なのだが、非常に抽象度の高いプログラミング言語であるため、一度扱えるようになれば、今までより高度なプログラミングが可能となる。
###プログラミング
何度も言うように、プログラミングは抽象化が大事だ。
要は、プログラミングは抽象化の水準を上げていくことで発展してきたということだ。
この歴史をたどっていこう。
プログラミングは以下のように抽象度を上げてきた。
アセンブリ言語→ラムダ計算→手続き型プログラミング→オブジェクト指向プログラミング→関数型プログラミング
###構造
数学や化学、プログラミングができるまで進化を遂げた人間だが、古代から変わらないものもある。
それは脳の属性だ。
人間の脳は至る所に構造を見つけだしているのだ。
難しい問題を簡単なものに分解して、合成しているのだ。
この考え方がプログラミングにだって現れている。
